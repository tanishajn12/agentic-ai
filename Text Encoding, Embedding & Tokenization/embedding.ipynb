```python
from sentence_transformers import SentenceTransformer
import numpy as np
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
documents = [
    "Eating fiber reduces heart disease risk.",
    "Fruits and vegetables lower cardiovascular disease chances",
    "Buying a new car improves driving comfort.",
    "Regular exercise enhances cardiovascular health."
]
```

```python
query = "How to reduce heart disease risk?"
```

```python
# Generate embeddings for the documents and the query
doc_embeddings = model.encode(documents)
query_embedding = model.encode(query)
```

```python
print("Document Embeddings Shape:", doc_embeddings.shape)
print("SentenceTransformer embedding shape:", np.array(query_embedding).shape)
```

```python
#Rank documents
def cosine_similarity(a,b):
    a = np.array(a, dtype=np.float32)
    b = np.array(b, dtype=np.float32)
    return np.dot(a,b) / (np.linalg.norm(a) * np.linalg.norm(b))


scores = [cosine_similarity(query_embedding, doc_emb) for doc_emb in doc_embeddings]
ranked = sorted(zip(documents, scores), key=lambda x: x[1], reverse=True)

print("\nRanked Documents:")
for doc, score in ranked:
    print(f" - {doc} (Score: {score:.4f})")
```

```python
def cosine_similarity(a,b):
    a = np.array(a, dtype=np.float32)
    b = np.array(b, dtype=np.float32)
    return np.dot(a,b) / (np.linalg.norm(a) * np.linalg.norm(b))

def dot_product(a,b):
    a = np.array(a, dtype=np.float32)
    b = np.array(b, dtype=np.float32)
    return np.dot(a,b)

def euclidean_distance(a,b):
    a = np.array(a, dtype=np.float32)
    b = np.array(b, dtype=np.float32)
    return np.linalg.norm(a - b)
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np

#load model
model = SentenceTransformer('all-MiniLM-L6-v2')


#data
documents = [
    "Eating fiber reduces heart disease risk.",
    "Fruits and vegetables lower cardiovascular disease chances",
    "Buying a new car improves driving comfort.",
    "Regular exercise enhances cardiovascular health."
]

query = "How to reduce heart disease risk?"

# Generate embeddings for the documents and the query
doc_embeddings = model.encode(documents)
query_embedding = model.encode(query)

results = []
for doc, doc_emb in zip(documents, doc_embeddings):
    cos_sim = cosine_similarity(query_embedding, doc_emb)
    dot_prod = dot_product(query_embedding, doc_emb)
    euc_dist = euclidean_distance(query_embedding, doc_emb)
    results.append({
        "document": doc,
        "cosine_similarity": cos_sim,
        "dot_product": dot_prod,
        "euclidean_distance": euc_dist
    })
    
    
print("Cosine Similarity Results: (Higher is better)")
for item in sorted(results, key=lambda x: x['cosine_similarity'], reverse=True):
    print(f" - {item['document']} (Cosine Similarity: {item['cosine_similarity']:.4f})")

print("\nDot Product Results: (Higher is better)")
for item in sorted(results, key=lambda x: x['dot_product'], reverse=True):
    print(f" - {item['document']} (Dot Product: {item['dot_product']:.4f})")
    
print("\nEuclidean Distance Results: (Lower is better)")
for item in sorted(results, key=lambda x: x['euclidean_distance']):
    print(f" - {item['document']} (Euclidean Distance: {item['euclidean_distance']:.4f})")
```
